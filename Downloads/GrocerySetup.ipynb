```python
import arcpy
import os

aprx = arcpy.mp.ArcGISProject("CURRENT")

base_dir = aprx.homeFolder

output_gdb = os.path.join(base_dir, "Grocery.gdb")

if not os.path.exists(output_gdb):
    arcpy.management.CreateFileGDB(base_dir, "Grocery.gdb")

m = aprx.listMaps("GEOG392 map")[0]

shops_layer = m.listLayers("OSM_NA_Shops")[0]

sql_exp = "shop IN ('supermarket', 'convenience')"

arcpy.management.Delete("shop_lyr")

arcpy.management.MakeFeatureLayer(shops_layer, "shop_lyr", sql_exp)

output_fc = output_gdb + "\GroceryStores"
if(arcpy.Exists(output_fc)):
    arcpy.management.Delete(output_fc)
    
arcpy.management.CopyFeatures("shop_lyr", output_fc)

print("Grocery stores exported successfully!")
print(output_fc)
```
